```python
import DataLoader as dl

# change the files accordingly
product_name = "Fashion"
meta_data_path = 'meta_AMAZON_FASHION.json.gz'
review_data_path = 'AMAZON_FASHION.json.gz'

# columns that we decided to use from review_data
review_features = ['reviewerID', 'asin','vote','reviewText','overall','summary','reviewTime']

# columns that we decided to use from meta_data
meta_features = [
    'asin','title', 'feature',
    'description', 'price', 'imUrl',
    'also_buy', 'also_bought', 'also_view', 'salesRank',
    'brand', 'categories',
]

# combine all to list so that we we use it to filter out columns that we want
desired_features = meta_features + review_features

data  = dl.load_data(
    review_data_path = review_data_path,
    meta_data_path = meta_data_path,
    desired_features = desired_features,
    product_name = product_name)

# explore null values
dl.explore_null(data)
```
